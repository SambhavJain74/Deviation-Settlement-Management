```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def select_columns_by_index(input_file, column_indexes, skiprows, nrows):
    df = pd.read_excel(input_file, usecols=column_indexes, dtype=str, skiprows=skiprows, nrows=nrows)
    return df

def process_sheet_data(input_file, sheet_name, selected_column_indexes, skiprows, nrows):
    df = select_columns_by_index(input_file, selected_column_indexes, skiprows, nrows)
    
    df = df.apply(pd.to_numeric, errors='coerce')
    df.fillna(df.mean(), inplace=True)

    df['Over-Injection upto 10% or 100MW'] = np.minimum(df.iloc[:, 0], 100)
    df['Over-Injection > 10% or 100MW'] = np.maximum(df.iloc[:, 0] - 100, 0)
    df['Under-Injection upto 10% or 100MW'] = np.minimum(df.iloc[:, 0], 100)
    df['Under-Injection > 10% or 100MW'] = np.maximum(df.iloc[:, 0] - 100, 0)
    
    df['Deviation Charges for over-injection (Rs.)'] = df['Over-Injection > 10% or 100MW'] * 3.3047
    df['Deviation Charges for under-injection (Rs.)'] = df['Under-Injection > 10% or 100MW'] * 3.3047
    
    df['Total Deviation Charges (Rs.)'] = df['Deviation Charges for over-injection (Rs.)'] + df['Deviation Charges for under-injection (Rs.)']
    
    mapped_output_file = "mapped_output.xlsx"
    df.to_excel(mapped_output_file, sheet_name=sheet_name, index=False)
    
    print(f"\nExploratory Data Analysis for {sheet_name}")
    
    print("\nSummary Statistics:")
    print(df.describe())
    
    plt.figure(figsize=(12, 6))
    for col in ["Over-Injection upto 10% or 100MW", "Over-Injection > 10% or 100MW",
                "Under-Injection upto 10% or 100MW", "Under-Injection > 10% or 100MW"]:
        sns.histplot(data=df, x=col, kde=True, label=col)
    plt.title("Distribution of Over-Injection and Under-Injection")
    plt.xlabel("Injection Amount (MW)")
    plt.ylabel("Frequency")
    plt.legend()
    plt.show()

    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df[['Deviation Charges for over-injection (Rs.)', 'Deviation Charges for under-injection (Rs.)']])
    plt.title("Deviation Charges Distribution")
    plt.ylabel("Deviation Charges (Rs.)")
    plt.show()

    plt.figure(figsize=(8, 5))
    sns.histplot(data=df, x='Total Deviation Charges (Rs.)', kde=True)
    plt.title("Distribution of Total Deviation Charges")
    plt.xlabel("Total Deviation Charges (Rs.)")
    plt.ylabel("Frequency")
    plt.show()

excel_file_name = "edited.xlsx"
skiprows = 6
endrows = 103
nrows = endrows - skiprows + 1
selected_column_indexes = [2, 3, 5, 7, 9, 12, 15]

excel_data = pd.read_excel(excel_file_name, sheet_name=None)
for sheet_name, _ in excel_data.items():
    process_sheet_data(excel_file_name, sheet_name, selected_column_indexes, skiprows, nrows)
```
